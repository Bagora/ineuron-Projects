Iteration 1:

2. Calculate the new centroids for each cluster:
   - New centroid for Cluster 1: (5 + 10 + 15) / 3 = 10
   - New centroid for Cluster 2: (20 + 25 + 30 + 35) / 4 = 27.5

Iteration 2:
1. Reassign data points to the nearest centroids:
   - Cluster 1: [5, 10, 15]
   - Cluster 2: [20, 25, 30, 35]

2. Calculate new centroids (which remain the same in this case).

Since there is no change in data point assignments in Iteration 2, the clustering is stable. The final clusters are the same as in Iteration 1:

Cluster 1: [5, 10, 15]
Cluster 2: [20, 25, 30, 35]

b) Calculate the SSE for each set of centroids:

The SSE for a cluster is the sum of the squared distances between each data point in the cluster and its centroid. Let's calculate it for both sets of centroids:

SSE for Set 1:
Cluster 1:
- Data points: [5, 10, 15]
- Centroid: 10
- SSE = [(5 - 10)^2 + (10 - 10)^2 + (15 - 10)^2] = (25 + 0 + 25) = 50

Cluster 2:
- Data points: [20, 25, 30, 35]
- Centroid: 27.5
- SSE = [(20 - 27.5)^2 + (25 - 27.5)^2 + (30 - 27.5)^2 + (35 - 27.5)^2] = (56.25 + 6.25 + 6.25 + 56.25) = 125

Total SSE for Set 1 = 50 + 125 = 175

SSE for Set 2:
Cluster 1 (same as Set 1):
- SSE = 50

Cluster 2 (same as Set 1):
- SSE = 125

Total SSE for Set 2 = 50 + 125 = 175

Both sets of centroids have the same SSE, which is 175.

````python
import numpy as np

# Given data points
data_points = np.array([5, 10, 15, 20, 25, 30, 35])

# Number of clusters (k)
k = 2

# Initial centroids for Set 1 and Set 2
initial_centroids_set1 = np.array([15, 32])
initial_centroids_set2 = np.array([12, 30])

def k_means_clustering(data, centroids):
    clusters = {}
    for centroid in centroids:
        clusters[centroid] = []

    # Assign data points to the nearest centroid
    for point in data:
        nearest_centroid = centroids[np.argmin(np.abs(centroids - point))]
        clusters[nearest_centroid].append(point)

    return clusters

def calculate_sse(clusters):
    sse = 0
    for centroid, points in clusters.items():
        mean = np.mean(points)
        squared_diff = np.sum((points - mean) ** 2)
        sse += squared_diff
    return sse

# Perform k-means clustering for Set 1
clusters_set1 = k_means_clustering(data_points, initial_centroids_set1)

# Calculate SSE for Set 1
sse_set1 = calculate_sse(clusters_set1)

# Perform k-means clustering for Set 2
clusters_set2 = k_means_clustering(data_points, initial_centroids_set2)

# Calculate SSE for Set 2
sse_set2 = calculate_sse(clusters_set2)

print(f"SSE for Set 1: {sse_set1}")
print(f"SSE for Set 2: {sse_set2}")
````

### Ans 2

Market Basket Analysis (MBA) is a data mining technique that utilizes association analysis concepts to uncover valuable insights from transactional data, particularly in retail and e-commerce. It identifies associations and patterns in customer purchase behavior by examining which items are frequently bought together. Here's how MBA makes use of association analysis concepts:

1. **Association Rules**: MBA generates association rules, such as "if a customer buys A, they are likely to buy B." These rules provide actionable insights for cross-selling, product placement, and targeted marketing.

2. **Support**: The support metric measures how often itemsets occur together in transactions, indicating the popularity of certain item combinations. High support suggests strong associations.

3. **Confidence**: Confidence quantifies the likelihood that when customers buy one item, they will also buy another. Higher confidence indicates stronger associations between items.

4. **Lift**: Lift measures the strength of association between itemsets while accounting for the expected occurrence. Lift values greater than 1 indicate a positive association.

By applying association analysis concepts, MBA helps businesses optimize sales strategies, improve customer satisfaction, and enhance inventory management by understanding buying patterns and preferences.

### Ans 3

The Apriori algorithm is a classic approach for learning association rules from transactional data. Let's consider a retail store as an example:

**Step 1: Data Collection**
Collect transaction data that records customer purchases over time. Each transaction is a list of items bought by a customer.

**Step 2: Data Preprocessing**
Organize the transaction data into a suitable format, such as a binary matrix where rows represent transactions and columns represent items. Set the values in the matrix to 1 if an item is present in a transaction and 0 otherwise.

**Step 3: Set Minimum Support and Confidence Thresholds**
Define minimum support and confidence thresholds. For example, you may set a minimum support of 0.2 (meaning an itemset must appear in at least 20% of transactions) and a minimum confidence of 0.6 (indicating a rule must be at least 60% accurate).

**Step 4: Generate Frequent Itemsets**
Initially, find frequent itemsets (item combinations) that meet the minimum support threshold. Start with individual items and progressively combine them into larger itemsets until no more frequent itemsets can be found. For example, if {A}, {B}, and {C} are frequent, then {A, B}, {A, C}, and {B, C} will also be checked.

**Step 5: Generate Association Rules**
For each frequent itemset, generate association rules based on the minimum confidence threshold. For example, if {A, B} is a frequent itemset, you can generate the rule {A} => {B} with confidence calculated as (support({A, B}) / support({A})).

**Step 6: Evaluate Rules**
Evaluate the generated association rules based on support, confidence, and lift (a measure of rule significance). Select the most interesting and actionable rules for business decisions.

For instance, the Apriori algorithm can reveal rules like "If customers buy milk and bread, they are 70% likely to purchase eggs as well." These rules help businesses optimize product placement, promotions, and recommendations to improve sales and customer satisfaction.

### Ans 4

In hierarchical clustering, the distance between clusters is measured using various linkage methods, which define how the distance between clusters is computed. The commonly used linkage methods include:

1. **Single Linkage (Minimum Linkage)**: It measures the distance between the closest (most similar) data points in two clusters. The distance between clusters is defined by the minimum pairwise distance between the data points of the two clusters.

2. **Complete Linkage (Maximum Linkage)**: It measures the distance between the farthest (most dissimilar) data points in two clusters. The distance between clusters is defined by the maximum pairwise distance between the data points of the two clusters.

3. **Average Linkage**: It calculates the average distance between all possible pairs of data points, one from each cluster.

4. **Centroid Linkage**: It calculates the distance between the centroids (mean points) of two clusters.

The choice of linkage method can significantly impact the resulting hierarchical clustering structure.

To decide when to end the iteration, hierarchical clustering uses a stopping criterion. Common criteria include:

- A predetermined number of clusters (cutting the dendrogram at a specific height).
- A specific distance threshold (clusters merge until the distance exceeds the threshold).
- A desired number of clusters or cluster sizes.
- Maximizing or minimizing a clustering quality metric (e.g., silhouette score or within-cluster variance).

The choice of stopping criterion depends on the problem and the desired clustering outcome.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.datasets import make_blobs

# Generate synthetic data
data, _ = make_blobs(n_samples=300, centers=4, random_state=0, cluster_std=0.60)

# Calculate linkage matrices for single-linkage, complete-linkage, and average-linkage
single_linkage = linkage(data, method='single')
complete_linkage = linkage(data, method='complete')
average_linkage = linkage(data, method='average')

# Plot dendrograms for all three linkage methods
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.title('Single-Linkage Dendrogram')
dendrogram(single_linkage, p=4, truncate_mode='level')

plt.subplot(132)
plt.title('Complete-Linkage Dendrogram')
dendrogram(complete_linkage, p=4, truncate_mode='level')

plt.subplot(133)
plt.title('Average-Linkage Dendrogram')
dendrogram(average_linkage, p=4, truncate_mode='level')

plt.tight_layout()
plt.show()
````

### Ans 5

In the k-means algorithm, cluster centroids are recomputed in an iterative manner as follows:

1. **Initialization**: Start by randomly selecting k initial centroids, where k is the number of clusters.

2. **Assignment Step**: For each data point in the dataset, calculate its distance to each of the k centroids and assign it to the cluster associated with the nearest centroid.

3. **Update Step**: After all data points have been assigned to clusters, compute the new centroids for each cluster. This is done by taking the mean of all data points in that cluster. The centroid represents the center or average position of data points in the cluster.

4. **Repeat**: Repeat the assignment and update steps until a stopping criterion is met. Common criteria include a maximum number of iterations or until the centroids no longer change significantly between iterations.

5. **Convergence**: The algorithm converges when the centroids stabilize, meaning that they remain relatively unchanged between iterations or the assignment of data points to clusters no longer changes.

The process continues until convergence, resulting in clusters with well-defined centroids.

In this code:

1. Synthetic data with 4 clusters is generated.
2. The KMeans model is initialized with 4 clusters.
3. The fit method is used to fit the KMeans model to the data, which performs clustering.
4. Cluster centroids are obtained using cluster_centers_.
5. Data points are assigned to clusters using labels_.
Finally, data points and cluster centroids are plotted to visualize the clusters.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 clusters
data, _ = make_blobs(n_samples=300, centers=4, random_state=0, cluster_std=0.60)

# Initialize KMeans with 4 clusters
kmeans = KMeans(n_clusters=4)

# Fit the KMeans model to the data
kmeans.fit(data)

# Get cluster centroids
centroids = kmeans.cluster_centers_

# Get cluster assignments for each data point
labels = kmeans.labels_

# Plot data points and centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.title('K-Means Clustering')
plt.show()
````

### Ans 6

Determining the appropriate number of clusters is crucial at the start of a clustering exercise. One widely used method for this is the **Elbow Method**:

1. **K-Means Algorithm**: Begin by applying the K-means clustering algorithm with a range of k values (the number of clusters) on the dataset.

2. **Calculate Within-Cluster Sum of Squares (WCSS)**: For each k, calculate the sum of squared distances between data points and their assigned cluster centroids. This is called the WCSS.

3. **Plot the Elbow Curve**: Create a plot of k values against their corresponding WCSS values. As k increases, WCSS typically decreases because clusters become smaller and data points are closer to their centroids. The goal is to find the "elbow point" on the curve.

4. **Identify the Elbow Point**: The elbow point is where the rate of WCSS decrease sharply changes. It represents a trade-off between having too few clusters (underfitting) and too many clusters (overfitting). The k value at the elbow point is often a good choice for the number of clusters.

By visually inspecting the elbow curve, data analysts can make an informed decision about the appropriate number of clusters that best captures the underlying structure of the data. However, it's important to note that in some cases, the elbow method may not yield a clear elbow point, and domain knowledge or other validation techniques may be necessary.

In this code:

1. Synthetic data with 4 clusters is generated.
2. A loop iterates through a range of k values from 1 to 10.
3. For each k, K-means clustering is performed, and the WCSS (inertia) is calculated and appended to the wcss list.
4. The Elbow Curve is plotted to visualize the WCSS values for different k values. The "elbow point" represents the optimal number of clusters.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 clusters
data, _ = make_blobs(n_samples=300, centers=4, random_state=0, cluster_std=0.60)

# Initialize an empty list to store WCSS values
wcss = []

# Try a range of k values (number of clusters)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(data)
    # Calculate WCSS for each k and append to the list
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid()
plt.show()
````

### Ans 7

**Advantages of K-Means Algorithm:**

1. **Simple and Efficient:** K-means is easy to implement and computationally efficient, making it suitable for large datasets.

2. **Scalability:** It can handle a large number of data points and dimensions.

3. **Clustering Quality:** Works well when clusters are spherical, equally sized, and have similar densities.

**Disadvantages of K-Means Algorithm:**

1. **Sensitive to Initial Centroids:** Results may vary depending on the initial centroids, leading to suboptimal solutions.

2. **Assumes Spherical Clusters:** Struggles with non-spherical or irregularly shaped clusters.

3. **Requires Predefined K:** The number of clusters (K) must be specified in advance, which can be challenging.

4. **Sensitive to Outliers:** Outliers can significantly impact cluster assignments.

In this code:

1. Synthetic data with 3 clusters is generated.
2. K-means clustering is applied to the data using scikit-learn.
3. The advantages and disadvantages of the K-means algorithm are discussed in the comments.

````python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data with 3 clusters
data, _ = make_blobs(n_samples=300, centers=3, random_state=0, cluster_std=0.60)

# Initialize KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, n_init=10)

# Fit KMeans to the data
kmeans.fit(data)

# Get cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot data points and centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.title('K-Means Clustering')
plt.show()

# Discuss the advantages and disadvantages
advantages = [
    "Simple and efficient algorithm",
    "Scalable to large datasets",
    "Efficient for well-separated spherical clusters",
]

disadvantages = [
    "Sensitive to initial centroid selection",
    "Assumes spherical clusters, struggles with irregular shapes",
    "Requires predefined number of clusters (K)",
    "Sensitive to outliers and noise",
]

print("Advantages of K-Means:")
for adv in advantages:
    print(f"- {adv}")

print("\nDisadvantages of K-Means:")
for disadv in disadvantages:
    print(f"- {disadv}")
````
