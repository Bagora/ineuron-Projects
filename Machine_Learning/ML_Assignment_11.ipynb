````python
import numpy as np
from scipy.stats import norm

# Define the PMF for X_i (you'll need to provide your specific PMF)
# For this example, let's assume X_i follows a discrete uniform distribution
# You should replace this with your actual PMF.
pmf_X = {1: 1/6, 2: 1/6, 3: 1/6, 4: 1/6, 5: 1/6, 6: 1/6}

# Number of random variables
n = 25

# Calculate the mean (mu_X) and standard deviation (sigma_X) for X_i
mu_X = sum(x * p for x, p in pmf_X.items())
sigma_X = np.sqrt(sum((x - mu_X)**2 * p for x, p in pmf_X.items()))

# Calculate the mean (mu_Y) and standard deviation (sigma_Y) for Y
mu_Y = n * mu_X
sigma_Y = np.sqrt(n) * sigma_X

# Define the range for Y
lower_bound = 4
upper_bound = 6

# Standardize the bounds using Z-scores
z_lower = (lower_bound - mu_Y) / sigma_Y
z_upper = (upper_bound - mu_Y) / sigma_Y

# Calculate the probability P(4 <= Y <= 6) using the standard normal distribution
probability = norm.cdf(z_upper) - norm.cdf(z_lower)

print("Estimated probability P(4 <= Y <= 6):", probability)
````
