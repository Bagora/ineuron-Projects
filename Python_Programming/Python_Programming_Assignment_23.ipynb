1. Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not. A number is symmetrical when it is the same as its reverse.
    Examples
    is_symmetrical(7227) ➞ True

    is_symmetrical(12567) ➞ False

    is_symmetrical(44444444) ➞ True

    is_symmetrical(9939) ➞ False

    is_symmetrical(1112111) ➞ True

2. Given a string of numbers separated by a comma and space, return the product of the numbers.
    Examples
    multiply_nums("2, 3") ➞ 6

    multiply_nums("1, 2, 3, 4") ➞ 24

    multiply_nums("54, 75, 453, 0") ➞ 0

    multiply_nums("10, -2") ➞ -20

3. Create a function that squares every digit of a number.
    Examples
    square_digits(9119) ➞ 811181

    square_digits(2483) ➞ 416649

    square_digits(3212) ➞ 9414
    Notes
    The function receives an integer and must return an integer.

4. Create a function that sorts a list and removes all duplicate items from it.
    Examples
    setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5]

    setify([4, 4, 4, 4]) ➞ [4]

    setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15]

    setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]

5. Create a function that returns the mean of all digits.
    Examples
    mean(42) ➞ 3

    mean(12345) ➞ 3

    mean(666) ➞ 6
    Notes
    •	The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in 512 is (5+1+2)/3(number of digits) = 8/3=2).
    •	The mean will always be an integer.

```python
'''Ans 1:- We convert the input number num to a string using str() so
that we can easily reverse it and compare it with its original
form.  We use string slicing with [::-1] to reverse the string
representation of the number.  Finally, we compare the original string
with its reverse. If they are equal, it means the number is
symmetrical, and we return True. Otherwise, we return False.'''

def is_symmetrical(num):
    # Convert the number to a string to work with its digits
    num_str = str(num)
    
    # Check if the string is equal to its reverse
    return num_str == num_str[::-1]

# Test cases
print(is_symmetrical(7227))
print(is_symmetrical(12567))
print(is_symmetrical(44444444))
print(is_symmetrical(9939))
print(is_symmetrical(1112111))
```

```python
'''Ans 2:- The function product_of_numbers we use a while True loop
to continuously ask for input until the user enters 'q'. 
Inside the loop, we use the input() function to get a string of
numbers from the user.  We check if the input is 'q', and if
so, we break out of the loop to exit the program.  Otherwise,
we call the product_of_numbers function with the input
string, calculate the product, and print the result.  The loop
continues until the user decides to quit by entering 'q'.'''

def product_of_numbers(numbers_str):
    numbers = numbers_str.split(", ")
    product = 1
    
    for num_str in numbers:
        try:
            num = float(num_str)
            product *= num
        except ValueError:
            return "Invalid input: Please provide a list of numbers separated by commas and spaces."
    
    return product

while True:
    input_str = input("Enter a list of numbers separated by commas and spaces (or 'q' to exit): ")
    
    if input_str.lower() == 'q':
        break
    
    result = product_of_numbers(input_str)
    print("Product:", result)
```

```python
'''Ans 3:- The square_digits function takes a number as input.  It
converts the number to a string to iterate through its digits.  It
initializes an empty string called squared_str to store the squared
digits.  It iterates through the digits in the string, converts
each digit string to an integer, squares it, and adds the
squared digit back to the squared_str.  Finally, it converts the
result string back to an integer and returns it.  In the main
part of the code, a while True loop continuously asks the user
for input until they enter 'quit'. It then calls the
square_digits function to calculate and display the squared digits of
the entered number.'''

def square_digits(number):
    num_str = str(number)
    
    # Initialize an empty string to store the squared digits
    squared_str = ""
    
    # Iterate through the digits in the string
    for digit_str in num_str:
        try:
            digit = int(digit_str)
            squared_digit = digit ** 2
            squared_str += str(squared_digit)
        except ValueError:
            return "Invalid input: Please provide a valid number."
    
    return int(squared_str)

while True:
    user_input = input("Enter a number (or 'q' to exit): ")
    
    if user_input.lower() == 'q':
        break
    
    try:
        number = int(user_input)
        result = square_digits(number)
        print("Squared digits:", result)
    except ValueError:
        print("Invalid input: Please provide a valid number.")
```

```python
'''Ans 4:- The sort_and_remove_duplicates function takes a list as
input, sorts it, and removes duplicate items.  It initializes an
empty list called unique_list to store unique items.  It
iterates through the sorted list and adds items to unique_list only
if they are not already present.  In the main part of the
code, a while True loop continuously asks the user for input
until they enter 'quit'. It then splits the input string into a
list of items, calls the sort_and_remove_duplicates function to
process the list, and displays the sorted and duplicate-free list.'''

def sort_and_remove_duplicates(input_list):
    # Sort the list
    sorted_list = sorted(input_list)
    
    # Initialize an empty list to store unique items
    unique_list = []
    
    # Iterate through the sorted list and add unique items to unique_list
    for item in sorted_list:
        if item not in unique_list:
            unique_list.append(item)
    
    return unique_list

while True:
    user_input = input("Enter a list of numbers separated by commas (or 'quit' to exit): ")
    
    if user_input.lower() == 'quit':
        break
    
    input_list = user_input.split(", ")  # Split the input string into a list of items
    input_list = [int(item) for item in input_list]  # Convert items to integers
    result = sort_and_remove_duplicates(input_list)
    
    # Print the result as a list of numbers
    print("Sorted and duplicate-free list:", result)
```

```python
'''Ans 5:- We define a single function calculate_mean_of_digits that
takes the number as input and calculates the mean of its digits.
Inside the function, we convert the number to a string,
extract its digits as integers using a list comprehension, and
calculate the mean.  We handle invalid inputs using a try-except
block and return an error message if needed.  The while True
loop allows the user to enter numbers repeatedly until they
enter 'q', and it calculates and displays the mean of digits
for each input.'''

def calculate_mean_of_digits(number):
    try:
        digits = [int(digit) for digit in str(number)]
        return sum(digits) / len(digits)
    except ValueError:
        return "Invalid input: Please provide a valid number."

while True:
    user_input = input("Enter a number (or 'q' to exit): ")
    
    if user_input.lower() == 'q':
        break
    
    try:
        number = int(user_input)
        print(f"Mean of digits:- {calculate_mean_of_digits(number)}")
    except ValueError:
        print("Invalid input: Please provide a valid number.")
```
