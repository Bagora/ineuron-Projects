```python
'''Ans 1:- Regular expressions (regex) offer several benefits when working with text data
and pattern matching in programming:-

1. Pattern Matching: Regular expressions provide a powerful and flexible way to
search for and match patterns within strings. They allow you to define complex search
patterns using a compact syntax.

2. Text Manipulation: Regular expressions enable you to perform text manipulation
tasks such as searching, replacing, and extracting specific parts of strings with
ease. This is particularly useful for tasks like data cleaning and transformation.

3. Efficiency: Regular expressions are optimized for efficient pattern matching.
They often outperform manual string manipulation methods when dealing with complex
patterns.

4. Flexibility: Regular expressions support a wide range of patterns and rules,
including character classes, quantifiers, alternatives, and more. This makes them
suitable for handling various text processing scenarios.

5. Cross-Language Support: Regular expressions are supported in many programming
languages, not just Python. This makes it easier to apply your knowledge across different
projects and languages.

6. Validation: Regular expressions are commonly used for data validation tasks,
such as checking if an email address or phone number follows a specific format.

7. Shorter Code: Regular expressions can often replace multiple lines of code
with a single expression, making code more concise and easier to read.

8. Advanced Search and Replace: Regular expressions allow you to perform complex
search and replace operations, such as finding all occurrences of a pattern and
replacing them in a specific way.

Despite their benefits, regular expressions can also be complex and hard to
read for complex patterns. Learning and understanding regex syntax takes time, but
the skills gained can significantly improve your ability to handle text-related
tasks efficiently and effectively.

In this example, the regular expression
r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b' is used to match email addresses.
The pattern breaks down as follows:-

\b: Word boundary to ensure the email is not part of a larger word.
[A-Za-z0-9._%+-]+: Matches the username part before the "@" symbol.
@: Matches the "@" symbol.
[A-Za-z0-9.-]+: Matches the domain name.
\.: Matches the dot before the top-level domain.
[A-Z|a-z]{2,}: Matches the top-level domain with at least 2 characters.
\b: Word boundary to ensure the email is not part of a larger word.

The re.findall() function searches the text for all occurrences of the pattern
and returns a list of matched email addresses.  Regular expressions provide a
powerful way to define and search for complex patterns like email addresses, making
tasks like data extraction and validation much more efficient.'''

import re

text = "Contact us at mark@example.com or don@test.com for inquiries."

# Define a regular expression pattern to match email addresses
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

# Find all occurrences of email addresses in the text
email_addresses = re.findall(email_pattern, text)

print(email_addresses)
```

```python
'''Ans 2:- The regular expressions "(ab)c+" and "a(bc)+" have different effects when used
for pattern matching:-

1. "(ab)c+": This pattern matches sequences that start with "ab" 
It captures the entire "ab" sequence as a group and then matches.

2. "a(bc)+": This pattern matches sequences that start with "a" followed by one
or more occurrences of the sequence "bc". It captures the entire "bc" sequence as
a group and then matches one or more occurrences of "bc".

The unqualified pattern "abc+" refers to the pattern that matches one or more
occurrences of the sequence "abc". In this case, "abc+" would match strings like "abc",
"abccc", "abcccc", and so on.

In summary:-
"(ab)c+": Matches sequences of "ab" characters.
"a(bc)+": Matches sequences of "a" followed by one or more occurrences of "bc".
"abc+": Matches sequences of "abc" with one or more occurrences of "c".

Here's an example to illustrate the differences between the regular
expressions "(ab)c+" and "a(bc)+" using the re module in Python:

In this example, the input text contains various sequences. The patterns
"(ab)c+", "a(bc)+", and "abc+" are used to find matches in the text. The outputs
demonstrate the different matches produced by each pattern.'''

import re

text = "abcc abcbc abcbcbc abc abccc"

# Pattern "(ab)c+"
pattern1 = r"(ab)c+"
matches1 = re.findall(pattern1, text)

# Pattern "a(bc)+"
pattern2 = r"a(bc)+"
matches2 = re.findall(pattern2, text)

# Pattern "abc+"
pattern3 = r"abc+"
matches3 = re.findall(pattern3, text)

print("Matches for (ab)c+:", matches1)
print("Matches for a(bc)+:", matches2)
print("Matches for abc+:", matches3)
```

```python
'''Ans 3:- The statement "import re" is necessary when using regular expressions in
Python as it allows you to access the functionality provided by the re module, which
is essential for working with regular expressions. The re module provides various
functions and methods for pattern matching and manipulation of strings using regular
expressions.

By importing the re module, you gain access to functions like re.match(),
re.search(), re.findall(), re.sub(), and more, which are used for performing operations
such as searching for patterns, extracting text, and replacing substrings based on
regular expressions.

In this example, the import re statement is crucial for utilizing the re
module's capabilities to search for and extract an email address using regular
expressions.  In summary, when working with regular expressions in Python, the "import re"
statement is a fundamental requirement to access the regular expression functions and
methods provided by the re module.'''

import re

text = "Hello, my email is user@example.com."

# Using re.search() to find an email address
pattern = r'\b[\w\.-]+@[\w\.-]+\.\w+\b'
match = re.search(pattern, text)

if match:
    print("Found email:", match.group())
else:
    print("No email found.")
```

```python
'''Ans 4:- In regular expressions, square brackets ([]) are used to define a character
class, which allows us to specify a set of characters that we want to match. Inside
square brackets, some characters have special significance and are used to define
character ranges or special character classes. 

1. Hyphen (-): The hyphen is used to define character ranges within square
brackets. For example, [a-z] represents all lowercase alphabetic characters from "a" to
"z", inclusive.

2. Caret (^): When the caret is the first character inside square brackets, it
negates the character class. It matches any character that is not in the specified
set. For example, [^0-9] matches any character that is not a digit.

3. Backslash (): The backslash is used to escape characters with special meaning
to treat them as literal characters. For example, [\+] matches the plus sign "+".

4. Dash and Caret ([-^]): To match a literal dash or caret character inside
square brackets, place it at the beginning or end of the character class. For
example, [-abc] matches a literal hyphen or the characters "a", "b", or "c".

Other characters inside square brackets typically have their literal meanings.
For example, [abc] matches any of the characters "a", "b", or "c".  Keep in mind
that these special characters have their meanings only within square brackets.
Outside square brackets, they don't usually have the same significance.

In this example, the character class [a-z0-9] matches any lowercase letter or
digit in the input text. The hyphen - is used to specify the range of lowercase
letters, and the digits 0-9 are explicitly listed.'''

import re

text = "a1b2c3"

# Matches any lowercase letter or digit using character classes
pattern = r'[a-z0-9]'
matches = re.findall(pattern, text)

print(matches)
```

```python
'''Ans 5:- Compiling a regular expression object in Python using the re.compile()
function offers several benefits that can enhance the performance and readability of
your code.

1. Improved Performance: Compiling a regular expression into an object allows
Python to optimize the pattern matching process. The compiled object can be reused
multiple times, leading to faster pattern matching operations, especially when the same
pattern is used frequently.

2. Readability: By compiling a regular expression separately, we can give it a
meaningful variable name, making our code more self-explanatory. This improves code
readability, as the purpose of the regular expression is clear at a glance.

3. Reduced Overhead: Compiling a regular expression object involves some initial
overhead, but the compiled object can be reused efficiently. This reduces the overhead
associated with parsing and processing the regular expression pattern each time it's
used.

4. Avoiding Redundant Compilation: In some cases, we might use the same regular
expression pattern multiple times in our code. Compiling it once into an object ensures
that we don't redundantly compile the same pattern in different places, saving
computational resources.

5. Changing Flags: Regular expression objects allow us to specify flags (such as
re.IGNORECASE for case-insensitive matching) when compiling. This lets you use the same
flags consistently across multiple uses of the pattern.

In this example, the regular expression pattern r'\d{3}-\d{2}-\d{4}' is
compiled into an object using re.compile(). The compiled pattern is then used for
searching the input text. This approach improves performance and readability.  In
summary, compiling a regular expression object offers performance optimization, better
readability, reduced overhead, and the ability to reuse patterns efficiently across your
codebase.'''

import re

# Compile a regular expression pattern into an object
pattern = re.compile(r'\d{3}-\d{2}-\d{4}')

# Use the compiled pattern for matching
text = "Social Security Number: 123-45-6789"
match = pattern.search(text)

if match:
    print("SSN found:", match.group())
else:
    print("No SSN found.")
```

```python
'''Ans 6:- The re.match() and re.search() functions in Python's re module return match
objects that provide information about the pattern matching process. Here are some
examples of how to use the match objects returned by these functions.

These examples showcase different ways to utilize match objects returned by
re.match() and re.search(). Match objects provide methods and attributes like group(),
start(), end(), and named groups to access matched text, indices, and captured groups,
making it easier to work with and process the results of pattern matching operations.'''

# Accessing Matched Text

import re

text = "Hello, world!"
pattern = r"Hello"
match = re.match(pattern, text)
if match:
    print("Matched text:", match.group())
    
    
# Accessing Matched Indices

text = "Hello, world!"
pattern = r"Hello"
match = re.search(pattern, text)
if match:
    print("Matched indices:", match.start(), match.end())
    

# Using Matched Text in Substitutions

text = "Hello, world!"
pattern = r"Hello"
replacement = "Hi"
new_text = re.sub(pattern, replacement, text)
print(new_text)


# Using Groups in Match Objects

text = "Name: John, Age: 30"
pattern = r"Name: (\w+), Age: (\d+)"
match = re.search(pattern, text)
if match:
    name = match.group(1)
    age = match.group(2)
    print("Name:", name)
    print("Age:", age)
    
# Using Named Groups in Match Objects

text = "Date: 2023-08-24"
pattern = r"Date: (?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})"
match = re.search(pattern, text)
if match:
    year = match.group("year")
    month = match.group("month")
    day = match.group("day")
    print("Year:", year)
    print("Month:", month)
    print("Day:", day)
```

```python
'''Ans 7:- The vertical bar (|) and square brackets ([]) serve different purposes in
regular expressions:-

1. Vertical Bar (|) - Alteration: The vertical bar is used to represent
alternation, allowing us to match one out of several possible alternatives. It is used to
create a logical "OR" condition between the patterns. For example, cat|dog matches
either "cat" or "dog".

2. Square Brackets ([]) - Character Set: Square brackets are used to define a
character set, allowing us to match any single character from the set. It's used to
specify a range of characters, and you're looking to match any one of them. For
example, [aeiou] matches any vowel.

In summary:  Vertical Bar (|) is used for alternation between multiple
patterns. Square Brackets ([]) define a character set to match any one character from
the set. While both constructs are used to specify different types of matching
conditions, they are not interchangeable. Use the vertical bar when we want to match
entire alternative patterns, and use square brackets when we want to match any one
character from a specific set of characters.'''

# Vertical Bar (|)

import re

text = "I have a cat and a dog."
pattern = r"cat|dog"
matches = re.findall(pattern, text)
print(matches)


# Square Brackets ([])

text = "The quick brown fox jumps over the lazy dog."
pattern = r"[aeiou]"
matches = re.findall(pattern, text)
print(matches)
```

```python
'''Ans 8:- In regular expression search patterns, using the raw-string indicator (r) is
necessary to prevent unwanted escaping of backslashes and special characters. Regular
expressions often contain backslashes to represent special sequences, and Python's string
literals also use backslashes for escape sequences. Without the raw-string indicator,
Python's interpretation of escape sequences could conflict with the escape sequences
used in regular expressions.

In replacement strings, using the raw-string indicator is not typically
necessary, as there are no escape sequences to worry about. However, using a raw string
can prevent accidental misinterpretation of backslashes in replacement strings
that might be used with functions like re.sub().

In summary, using the raw-string indicator (r) in regular expression search
patterns prevents conflicts between escape sequences, and using it in replacement
strings ensures backslashes are treated as literal characters, minimizing unintended
interpretations and errors.'''

import re

# Using raw-string indicator in search pattern

pattern = r"\d+"  # Matches one or more digits
text = "There are 123 apples and 456 oranges."
matches = re.findall(pattern, text)
print(matches)

# Using raw-string indicator in replacement string

pattern = r"\d+"
replacement = r"\\text"  # Use double backslash to represent a literal backslash
new_text = re.sub(pattern, replacement, text)
print(new_text)
```
