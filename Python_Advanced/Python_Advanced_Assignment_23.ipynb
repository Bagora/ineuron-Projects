```python
'''Ans1:- The code uses Matplotlib to create a comparison
between the sine and cosine functions. It generates an array of x
values ranging from 0 to 10, calculates the corresponding y
values for sine and cosine, and then plots both curves on the
same graph. The label parameter is used to label each curve,
and plt.legend() adds a legend. The title, axis labels, and
grid are set to provide context and readability, and plt.show()
displays the plot.

We also Have:-
Enhance graph comparisons: Use colors, legends,
annotations for clarity. Employ transparency, facets for overlaps.
Apply secondary axes for varied scales. Transform data, add
labels for precision. Keep axes consistent. Titles, captions
guide. Interactive elements engage. Small multiples aid
comprehension. Reference lines emphasize. Animation for time-based
insights. Adapt techniques to data, audience, and goals for
effective visualization.'''

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sin(x)')
plt.plot(x, y2, label='Cos(x)')
plt.legend()
plt.title('Sine and Cosine Comparison')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
'''Ans 2:- Compound interest yields higher returns over time compared
to a higher non-compounded interest rate. It reinvests earned
interest, accelerating growth. For example, a $1000 investment at 5%
annual interest compounded annually grows to $2700 approx in 20 years,
while the same investment at 7% non-compounded yields $2400.

This code illustrates that, even with a lower interest
rate, compound interest accumulates more due to reinvestment.
Compound interest benefits long-term investments, while a higher
non-compounded rate might initially seem attractive but falls short over
time.'''

principal = 1000
rate_compounded = 0.05
rate_non_compounded = 0.07
years = 20

# Compound interest formula
amount_compounded = principal * (1 + rate_compounded)**years

# Non-compounded interest formula
amount_non_compounded = principal * (1 + rate_non_compounded * years)

print("Compound Interest:", amount_compounded)
print("Non-Compounded Interest:", amount_non_compounded)
```

```python
'''Ans 3:- A histogram is a graphical representation of the
distribution of a dataset. It showcases how data is spread across
different bins or intervals. The x-axis represents the range of
values, divided into bins, while the y-axis shows the frequency or
count of data points falling within each bin. This visualization
is particularly useful for understanding the data's central
tendency, spread, and shape.

In NumPy, the numpy.histogram() method creates histograms.
It takes input data and bin specifications to compute
histogram values and bin edges. Here's an example:
In this example, the random data is used to create a
histogram with 20 bins. The resulting plot showcases the
distribution of the dataset, revealing patterns and insights about the
data's behavior.'''

import numpy as np
import matplotlib.pyplot as plt

# Generating random data
data = np.random.randn(1000)

# Compute histogram
hist_values, bin_edges = np.histogram(data, bins=20)  

# Create histogram plot
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
'''Ans 4:- We can adjust the aspect ratios between the X and Y axes
in a plot by using the plt.gca().set_aspect() method in
Matplotlib. This method allows us to set the aspect ratio to a
specific value or 'equal' for equal scaling. The aspect ratio can
influence how the data is visually represented on the plot.

In this example, the aspect ratio is set to 'equal', which
means that one unit on the X-axis will have the same physical
length as one unit on the Y-axis. This can be useful when we
want to ensure that the visual representation accurately
reflects the relationships between the variables, particularly in
cases where the scales of the X and Y axes are different.'''

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Change aspect ratio to 1 (equal scaling)
plt.gca().set_aspect('equal')
plt.show()
```

```python
'''Ans 5:- Dot product, outer product, and regular element-wise
multiplication are distinct operations when multiplying two NumPy arrays.

1. Dot Product: The dot product, calculated using np.dot() or
array's .dot() method, computes the sum of element-wise products.
It's used for matrix multiplication, requiring compatible
dimensions.

2. Outer Product: The outer product, generated with
np.outer(), calculates the cross-product of two arrays. It returns a
new array where each element is the product of one element
from the first array and all elements from the second.

3. Element-Wise Multiplication: This is the regular
multiplication of corresponding elements in two arrays using *. The
arrays must have matching shapes.

In summary, dot product aggregates element-wise products,
outer product constructs a new array from pairwise
multiplications, and element-wise multiplication simply multiplies
corresponding elements. The choice depends on desired outcomes and
dimensional compatibility.'''

import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

dot_product = np.dot(A, B)
outer_product = np.outer(A, B)
element_wise = A * B

print(f"Dot_product:- {dot_product} \nOuter_product:- {outer_product} \n Element_wise:- {element_wise }")
```

```python
'''Ans 6:- To estimate our monthly mortgage payment before buying a
home, We can use the numpy.pmt() function. This function
calculates the fixed monthly payment required to repay a loan (such
as a mortgage) based on its interest rate, loan duration, and
principal amount.

In this example, with a principal loan amount of $300,000,
an annual interest rate of 5%, and a loan duration of 30
years, the estimated monthly mortgage payment would be calculated
using the np.pmt() function. This helps us assess whether a
particular home purchase is financially feasible within our budget.

As np.pmt() function has been deprecated in after 1.20 NumPy
version. The error message suggests using the
numpy_financial library as a replacement. We can install this library
using: pip install numpy-financial'''

import numpy_financial as npf

principal = 300000
annual_interest_rate = 0.05
loan_duration_years = 30

monthly_interest_rate = annual_interest_rate / 12
loan_duration_months = loan_duration_years * 12

monthly_payment = npf.pmt(monthly_interest_rate, loan_duration_months, -principal)
print(f"Estimated Monthly Mortgage Payment:- ${monthly_payment}")
```
