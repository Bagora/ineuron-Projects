```python
'''Ans 1:- The PdfFileReader() and PdfFileWriter() objects themselves do not handle the
opening or closing of the file objects directly. Instead, you provide the file object
when creating these objects. The file object you provide should be opened in the
appropriate mode.  For PdfReader(), you generally want to open the PDF file in binary
read mode ('rb'). This is because PDF files contain binary data, and opening them
in text mode could lead to issues with character encoding.

For PdfWriter(), you also generally want to open the output PDF file in
binary write mode ('wb'), to ensure that the binary data is written correctly.
'''

from PyPDF2 import PdfReader, PdfWriter

# For reading a PDF file
pdf_reader = PdfReader(open('XGBoost.pdf', 'rb'))

# For writing to a PDF file
pdf_writer = PdfWriter()
with open('output.pdf', 'wb') as output_pdf:
    pdf_writer.write(output_pdf)
```

```python
'''Ans 2:-In this code, we're using the pages attribute of the PdfReader object to
access a specific page by its index (0-indexed). Then, we can perform various
operations on the Page object, such as extracting text, images, or modifying the content
of the page.

'''
from PyPDF2 import PdfReader

# Open the PDF file using PdfReader
pdf_reader = PdfReader('python.pdf')

# Get a Page object for page 5 (pages are 0-indexed)
page_number = 4  # Page 5 corresponds to index 4
page_object = pdf_reader.pages[page_number]

# Now you can work with the Page object
# For example, you can extract text from the page:
page_text = page_object.extract_text()
print(page_text)
```

```python
'''Ans 3:- The PdfFileReader variable doesn't directly store the number of pages in the
PDF document as an attribute. Instead, you can use the getNumPages() method of the
PdfFileReader class to retrieve the number of pages in the PDF document. Here's how you can
do it:
PdfFileReader is deprecated after 3.0, PdfReader object is the new one
'''
from PyPDF2 import PdfReader

# Open the PDF file
pdf_file_path = 'python.pdf'
pdf_file = open(pdf_file_path, 'rb')
pdf_reader = PdfReader(pdf_file)

# Get the number of pages using len(reader.pages)
num_pages = len(pdf_reader.pages)

# Close the PDF file
pdf_file.close()
print(f"Number of pages: {num_pages}")
```

```python
'''Ans 4:- If a PdfReader object's PDF is encrypted with a password then you need to
provide the password before you can obtain Page objects from it. You can do this by
using the decrypt(password) method of the PdfReader object. 

The pdf_reader.decrypt(password) line will decrypt the PDF using the provided
password before you can access the Page objects or other information from it.

First we need to encrypt a PDF file with a password using PyPDF2, by using
PdfWriter class to create a new encrypted PDF file with the desired password.

Here's a Python function that takes an input PDF file path, encrypts it with a
password, saves the encrypted PDF, then decrypts it using the same password and counts
the number of pages'''

from PyPDF2 import PdfReader, PdfWriter

def encrypt_and_decrypt_pdf(input_pdf_path, output_encrypted_pdf_path, password):
    # Encrypt the PDF
    pdf_reader = PdfReader(input_pdf_path)
    pdf_writer = PdfWriter()
    
    for page in pdf_reader.pages:
        pdf_writer.add_page(page)
    
    pdf_writer.encrypt(user_password=password, owner_pwd=None, use_128bit=True)
    
    with open(output_encrypted_pdf_path, 'wb') as output_pdf:
        pdf_writer.write(output_pdf)
    
    print(f"PDF encrypted and saved to {output_encrypted_pdf_path}")
    
    # Decrypt the encrypted PDF and count pages
    decrypted_pdf_reader = PdfReader(output_encrypted_pdf_path)
    decrypted_pdf_reader.decrypt(password)
    
    num_pages = len(decrypted_pdf_reader.pages)
    print(f"Decrypted PDF has {num_pages} pages")


input_pdf_path = 'XGBoost.pdf'
output_encrypted_pdf_path = 'encrypted_output.pdf'
password = "swordfish"

# Calling the function to get result
encrypt_and_decrypt_pdf(input_pdf_path, output_encrypted_pdf_path, password)
```

```python
'''Ans 5:- To rotate a page in a PDF using the PyPDF2 library, we can use the
rotateClockwise() and rotateCounterClockwise() methods of a Page object. These methods allow you
to rotate a page by 90 degrees clockwise or counterclockwise, respectively. 
'''
from PyPDF2 import PdfReader, PdfWriter

# Open the PDF file
pdf_file_path = 'python.pdf'
pdf_file = open(pdf_file_path, 'rb')
pdf_reader = PdfReader(pdf_file)

# Create a PdfWriter object
pdf_writer = PdfWriter()

# Rotate pages and add them to the writer
for page_number in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_number]
    rotated_page = page.rotate(90)  # Rotate by 90 degrees clockwise
    pdf_writer.add_page(rotated_page)

# Close the PDF file
pdf_file.close()

# Save the rotated PDF to a new file
output_pdf_path = 'rotated_output.pdf'
with open(output_pdf_path, 'wb') as output_pdf:
    pdf_writer.write(output_pdf)

print(f"PDF rotated and saved to {output_pdf_path}")
```

```python
'''Ans 7 :- If you have a Document object stored in a variable named doc and you want to
obtain a list of Paragraph objects from it, you can use the appropriate method
provided by the library you're using. Here's an example using the python-docx library
for working with Microsoft Word documents  In this example, the doc.paragraphs
attribute provides a list of Paragraph objects within the Document object. You can
iterate through this list to access and manipulate individual paragraphs as needed. 
If you're working with a different library or document format, the method to
obtain a list of Paragraph objects might vary. Always refer to the documentation of
the specific library you're using to understand the correct way to access
paragraphs from a Document object.'''

from docx import Document

# Load the document
doc = Document("Assignment_12.docx")  # Replace with the path to your document

# Obtain a list of Paragraph objects
paragraphs = doc.paragraphs

# Now the 'paragraphs' variable contains a list of Paragraph objects
for paragraph in paragraphs:
    print(paragraph.text)
```

```python
'''Ans 8: - In the context of document manipulation libraries like python-docx for
Microsoft Word documents, the attributes bold, underline, italic, strike, and outline
are typically associated with the "Run" objects. A "Run" object represents a
continuous span of characters within a paragraph that share the same formatting
attributes.  Here's an example of how you might use these attributes with python-docx:  In
this example, a paragraph is added to the document, and a run with formatted text
is added to the paragraph. The bold, italic, underline, strike, and outline
attributes of the run object are used to control the formatting of the text within that
run.  Remember that the usage of these attributes can vary based on the specific
library or framework you're working with. Always refer to the documentation of the
library to understand the correct usage and available attributes for formatting text
in your documents.'''

from docx import Document

doc = Document()

# Add a paragraph with a run
paragraph = doc.add_paragraph("This is a sample paragraph with ")
run = paragraph.add_run("formatted text.")
run.bold = True
run.italic = True
run.underline = True
run.strike = True
run.outline = True

# Save the document
doc.save("formatted_doc.docx")
```

```python
'''Ans 9:- In the context of working with text formatting attributes, such as the bold
attribute in document manipulation libraries like python-docx for Microsoft Word
documents, the values False, True, and None have specific meanings:  False:  When you set
the bold attribute of a run to False, you are indicating that the text within that
run should not be bold. The text will appear with the default or previously
applied formatting. True:  When you set the bold attribute of a run to True, you are
indicating that the text within that run should be bold. The text will appear with a bold
font weight. None:  When you don't explicitly set the bold attribute or set it to
None, the formatting will depend on the default or previous formatting applied. If
no bold formatting was applied previously, the text will appear without bold
formatting. If bold formatting was applied before, it will be retained. Here's an example
to illustrate the usage of these values in the context of python-docx. In this
example, the bold attribute is used to control the bold formatting of different runs
within the same paragraph. By setting it to True, False, or not setting it at all,
you can control whether the text appears bold or not in each run.'''

from docx import Document

doc = Document()
paragraph = doc.add_paragraph("This is a sample paragraph with ")
run = paragraph.add_run("formatted text.")

run.bold = True  # Set the text within this run to be bold

run2 = paragraph.add_run(" and this text is not bold.")  # Default formatting (not bold)

run3 = paragraph.add_run(" This text is bold again.")
run3.bold = True  # Set the text within this run to be bold again

doc.save("formatted_doc.docx")
```

```python
'''Ans 10 :-  Creating documents programmatically involves libraries like python-docx. The
Document Object Model (DOM) represents the structured hierarchy of a Word document.
Start by initializing a Document object as the container. Add content—text, tables,
images—sequentially. Control formatting for styling. Save the Document to a file for .docx output.
Libraries offer diverse features for automation, useful in report generation, invoicing,
and more. Always consult library docs for accurate usage details.'''

from docx import Document
from docx.shared import Inches

# Create a new document
doc = Document()

# Add content to the document
doc.add_heading('My New Document', level=1)
doc.add_paragraph('Python is everywhere!.')
doc.add_paragraph('AI is Future')

table = doc.add_table(rows=3, cols=3)
for row in table.rows:
    for cell in row.cells:
        cell.text = 'Data Science'

doc.add_picture('python-logo.png', width=Inches(4), height=Inches(4.5))

doc.save('my_document.docx')
```
