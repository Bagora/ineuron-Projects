```python
'''Ans 2:- In the csv module in Python, the csv.reader() and csv.writer() functions are
used to create reader and writer objects, respectively, for working with CSV
(Comma-Separated Values) files.

The csv.reader() function creates a reader object that allows
us to iterate through the rows of a CSV file. 

The csv.writer() function creates a writer object that allows us to write
data into a CSV file.

Remember to specify the appropriate file mode ('r' for reading or 'w' for
writing) when opening the file.'''

# csv.reader()
import csv

with open('data.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
        


# csv.writer()
with open('output.csv', 'w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['Name', 'Phone', 'email', 'address', 'Country', 'postalZip'])
    csv_writer.writerow(['Alice', '1-861-587-2298', 'ac@icloud.org' , 'P.O. Box 444, 3726 Eu Ave', 'USA' , '44423'])
    csv_writer.writerow(['Bob', '1-861-587-2298', 'ac@icloud.org' , 'P.O. Box 444, 3726 Eu Ave', 'Canada' , '44423'])

# The output.csv files now has this data
```

```python
'''Ans 3:- When using the csv.reader() function to create a reader object, you need to
open the file in read mode ('r'). This allows the reader object to read data from
the file.

When using the csv.writer() function to create a writer object, you
need to open the file in write mode ('w').

This allows the writer object to write data to the file. 
Make sure to use newline="" as well to handle newline
characters correctly across platforms.'''
```

```python
'''Ans 4:- The writerow() method is used to write the list data as a single row in the
CSV file. Each item in the list will correspond to a column in the CSV file. The
file will contain a row with the values 'Jon', 32, and 'North'.

In this example, first, we write the data to the CSV file using the csv_writer
object. Then, we open the file again in read mode and create a csv_reader object to
read the data. The for loop is used to iterate through the rows in the CSV file,
and each row will be a list of values representing the columns in that row.
'''

import csv

# Writing data to CSV
data = ['jon', 23, 'North']

with open('sample4.csv', 'w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(data)

# Reading data from CSV
with open('sample4.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
```

```python
'''Ans 5:- In the context of the csv module in Python, the keyword arguments delimiter
and line_terminator are used to customize how the CSV data is formatted when
writing and reading CSV files.

The delimiter keyword argument specifies the character
that separates individual fields (columns) within a row in a CSV file. By default,
a comma (,) is used as the delimiter.

The lineterminator keyword argument specifies the character or sequence of
characters that marks the end of a row in a CSV file. By default, a newline ('\n') or ('\r\n') is
used as the line terminator.


'''
import csv

data0 = ['Name', 'Age', 'country']
data1 = ['Alice', 25, 'USA']
data2 = ['Alice', 25, 'USA']

with open('sample5.csv', 'w', newline='') as file:
    csv_writer = csv.writer(file, delimiter='\t')
    csv_writer.writerow(data0)
    csv_writer.writerow(data1)
    csv_writer.writerow(data2)
    
# Reading the csv
import pandas as pd

sampledata = pd.read_csv("sample5.csv")
sampledata
```

```python
'''Ans 6:- The function that takes a string of JSON data and returns a Python data
structure is json.loads(). This function is part of the json module in Python and is
used to parse JSON (JavaScript Object Notation) strings and convert them into
native Python data structures.

In this example, the json_data string contains a JSON object. The json.loads()
function parses this JSON string and returns a corresponding Python dictionary. The
python_data variable will contain the parsed dictionary, which you can then work
with in your Python code.


'''

import json

json_data = '{"name": "Jon", "age": 25, "country": "North"}'
python_data = json.loads(json_data)
print(python_data)
```

```python
'''Ans 7:- The function that takes a Python data structure and returns a string of JSON
data is json.dumps(). This function is part of the json module in Python and is
used to serialize Python objects into JSON format.

In this example, the python_data dictionary is a Python data structure. The
json.dumps() function serializes this dictionary into a JSON-formatted string, which is
stored in the json_data variable. The printed output will be the JSON representation
of the dictionary.  You can also use the indent parameter to format the JSON data
with indentation for better readability:


'''

import json

python_data = {
    "name": "Mark",
    "age": 28,
    "country": "USA"
}

json_data = json.dumps(python_data)
print(json_data)
```
