```python
#Q1.What is the name of the feature responsible for generating Regex objects?
#Ans
'''
The re.compile() function returns Regex objects. We can combine a regular expression pattern into pattern objects, (re.Pattern)
which can be used for pattern matching. It also helps to search a pattern again without rewriting it
Later we can use this pattern object to search for a match inside different target strings using regex methods 
such as re.match() or re.search()
'''
import re

a = "My luck numbers are 281 711 231 441"
string_pattern = r"\d{2}" #pattern to find digits
regex_pattern = re.compile(string_pattern) #compile string pattern to re.Pattern object
result = regex_pattern.findall(a)
print('a = ', result)

b = "Your luck numbers are 115 216 514 995"
result = regex_pattern.findall(b)
print('b = ', result)
```

```python
#Q3.What is the return value of the search() method?
#Ans
'''
The search() method in Python's re (regular expression) module returns a Match object if a match is found, 
or None if no match is found.
The Match object contains information about the match, including the start and end indices of the matched substring,
as well as any groups that were defined in the regular expression pattern.
You can use the Match object to retrieve this information and use it in your code.

Here is an example of using the search() method to find a match in a string:
In this code, the regular expression \d+ is used to match one or more digits in a string.
The search() method is then used to find a match in the string "The number is 12345". 
If a match is found, the result variable will contain a Match object, which is used to print the matched substring.
If no match is found, the result variable will be None and the "No match found" message will be printed.
Overall, the search() method returns a Match object if a match is found, or None if no match is found.
This return value can be used to check if a match was found and to retrieve information about the match.
'''
import re

# Create a regular expression pattern
pattern = r"\d+"

# Use the search() method to find a match
result = re.search(pattern, "The number is 12345")

# Check if a match was found
if result:
    print(result.group())
else:
    print("No match found.")
```

```python
#Q4.From a Match item, how do you get the actual strings that match the pattern?
#Ans
'''
If you are using a regular expression, you can use the group() method to get the matched strings text. 
For example, suppose you have the following code:
This code will print "the" twice, since there are two occurrences of the word "the" in the input string.

'''
import re

string = "The quick brown fox jumps over the lazy dog"

# Find all occurrences of the word "the" in the string
pattern = re.compile(r"\bthe\b")
matches = pattern.finditer(string)

for match in matches:
    # Get the matched string
    matched_string = match.group()
    print(matched_string)
```

```python
'''
Q20.How would you write a regex that match a number with comma for every three digits? It must match the given
following: '42' '1,234' '6,368,745' but not the following: '12,34,567' (which has only two digits between the commas) '1234'
(which lacks commas)
'''
#Ans
'''
To match a number with a comma for every three digits, you could use the following regex:

re.compile(r'^\d{1,3}(,\d{3})*$')

Here is a breakdown of the regex:

r               # raw
^               # Anchor the regex to the start of the string
\d{1,3}         # Match one to three digits
(,\d{3})*       # Match zero or more groups of a comma followed by three digits
$               # Anchor the regex to the end of the string

This regex will match the following strings:

42
1,234
6,368,745

And it will not match the following strings:

12,34,567
1234


The first example, 12,34,567, does not match because it has only two digits between the commas. The second
example, 1234, does not match because it lacks commas.  Note that this regex is case-sensitive. It will only match numbers
that are written using the digits 0-9 but other regex strings can produce a similar regular expression. 
'''
```
